```python
#6.Write code to perform string operation using the count of repeated characters
def countChar(str, x):
    count = 0
    for i in range(len(str)): 
        if (str[i] == x) :
            count += 1
    n = 10
     
    # atleast k repetition are required 
    repetitions = n // len(str) 
    count = count * repetitions 
 
    # if n is not the multiple of the 
    # string size check for the remaining 
    
    l = n % len(str) 
    for i in range(l):
        if (str[i] == x): 
            count += 1

        return count 
str = "abcac"
print(countChar(str, 'a'))
 
```

```python
#7.write code to determine if string has unique characters
def has_unique_characters(s):
    """
    Check if a string has all unique characters.

    :param s: The string to check.
    :return: True if all characters in the string are unique, False otherwise.
    """
    return len(s) == len(set(s))

# Example usage
string_to_check = "hello"
print(has_unique_characters(string_to_check))
string_to_check = "abcde"
print(has_unique_characters(string_to_check))  
```

```python

def convert_case(s, to_upper=True):
    """
    Convert a given string to uppercase or lowercase.

    """
    if to_upper:
        return s.upper()
    else:
        return s.lower()


string_to_convert = "Hello, World!"

# Convert to uppercase
print(convert_case(string_to_convert, to_upper=True))  # Output: "HELLO, WORLD!"

# Convert to lowercase
print(convert_case(string_to_convert, to_upper=False))  # Output: "hello, world!"
```

```python
def count_words(s):
    """
    Count the number of words in a given string.

   
    """
    # Split the string into words using whitespace as the delimiter
    words = s.split()
    # Return the number of words
    return len(words)

# Example usage
string_to_count = "This is an example string with several words."
print(count_words(string_to_count))  # Output: 7
```

```python
def remove_element(lst, element_to_remove):
    # Create a new list with all occurrences of the element removed
    return [element for element in lst if element != element_to_remove]

# Example usage
my_list = [1, 2, 3, 4, 3, 5, 3]
element = 3
new_list = remove_element(my_list, element)
print(new_list)  # Output: [1, 2, 4, 5]
```

```python
def find_second_largest(nums):
    # Remove duplicates and check if there are at least two unique elements
    unique_nums = list(set(nums))
    if len(unique_nums) < 2:
        raise ValueError("List must contain at least two unique numbers")

    # Sort the unique numbers in descending order
    unique_nums.sort(reverse=True)

    # The second largest number is now the second element in the sorted list
    return unique_nums[1]

# Example usage
my_list = [10, 5, 20, 20, 4, 10]
second_largest = find_second_largest(my_list)
print(second_largest)  # Output: 10
```

```python
def count_occurrences(lst):
    # Create an empty dictionary to store the count of each element
    counts = {}
    
    # Iterate over each element in the list
    for element in lst:
        # If the element is already in the dictionary, increment its count
        if element in counts:
            counts[element] += 1
        # Otherwise, add the element to the dictionary with a count of 1
        else:
            counts[element] = 1
            
    return counts

# Example usage
my_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
occurrences = count_occurrences(my_list)
print(occurrences)  # Output: {1: 1, 2: 2, 3: 3, 4: 4}
```

```python
def reverse_list_in_place(lst):
    left = 0
    right = len(lst) - 1
    
    while left < right:
        # Swap the elements at the left and right indices
        lst[left], lst[right] = lst[right], lst[left]
        
        # Move the indices towards the center
        left += 1
        right -= 1

# Example usage
my_list = [1, 2, 3, 4, 5]
reverse_list_in_place(my_list)
```

```python
def remove_duplicates_preserve_order(lst):
    seen = set()
    result = []
    
    for element in lst:
        if element not in seen:
            seen.add(element)
            result.append(element)
    
    return result

# Example usage
my_list = [1, 2, 2, 3, 4, 4, 5, 1]
unique_list = remove_duplicates_preserve_order(my_list)
print(unique_list)  # Output: [1, 2, 3, 4, 5]
```

```python
def is_sorted(lst):
    # Check if the list is empty or has only one element
    if len(lst) < 2:
        return "List is sorted"
    
    # Initialize flags for ascending and descending order
    is_ascending = True
    is_descending = True

    # Iterate through the list to check the order
    for i in range(len(lst) - 1):
        if lst[i] < lst[i + 1]:
            is_descending = False
        elif lst[i] > lst[i + 1]:
            is_ascending = False

    # Determine the order based on the flags
    if is_ascending:
        return "List is sorted in ascending order"
    elif is_descending:
        return "List is sorted in descending order"
    else:
        return "List is not sorted"

# Example usage
list1 = [1, 2, 2, 3, 4]
list2 = [5, 4, 4, 2, 1]
list3 = [1, 3, 2, 4, 5]

print(is_sorted(list1))  # Output: List is sorted in ascending order
print(is_sorted(list2))  # Output: List is sorted in descending order
print(is_sorted(list3))  # Output: List is not sorted
```

```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0  # Pointers for list1 and list2
    
    # Traverse both lists and append the smaller element to merged_list
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    
    # Append any remaining elements from list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    
    # Append any remaining elements from list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    
    return merged_list

# Example usage
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]
```

```python
def intersection_of_lists(list1, list2):
    # Convert lists to sets to remove duplicates and use set operations
    set1 = set(list1)
    set2 = set(list2)
    
    # Find the intersection of both sets
    intersection = set1 & set2
    
    # Convert the result back to a list (if needed)
    return list(intersection)

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

intersection = intersection_of_lists(list1, list2)
print(intersection)  # Output: [3, 4, 5]
```

```python
def union_of_lists(list1, list2):
    # Convert lists to sets to remove duplicates and perform union
    set1 = set(list1)
    set2 = set(list2)
    
    # Find the union of both sets
    union = set1 | set2
    
    # Convert the result back to a list (if needed)
    return list(union)

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

union = union_of_lists(list1, list2)
print(union)  # Output: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
import random

def shuffle_list(lst):
    # Traverse the list in reverse
    for i in range(len(lst) - 1, 0, -1):
        # Pick a random index from 0 to i
        j = random.randint(0, i)
        
        # Swap the elements at index i and j
        lst[i], lst[j] = lst[j], lst[i]

# Example usage
my_list = [1, 2, 3, 4, 5]
shuffle_list(my_list)
print(my_list)  # Output: The list will be shuffled, e.g., [3, 1, 4, 5, 2]
```

```python
def common_elements(tuple1, tuple2):
    # Convert tuples to sets to find common elements
    set1 = set(tuple1)
    set2 = set(tuple2)
    
    # Find the intersection of both sets
    common_set = set1 & set2
    
    # Convert the result back to a tuple
    return tuple(common_set)

# Example usage
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (4, 5, 6, 7, 8)

common_tuple = common_elements(tuple1, tuple2)
print(common_tuple)  # Output: (4, 5)
```

```python
def get_string_set(prompt):
    # Prompt the user to enter a set of strings separated by commas
    user_input = input(prompt)
    # Split the input string by commas and convert each element to a string
    string_set = set(user_input.split(','))
    return string_set

def main():
    # Get two sets of strings from the user
    set1 = get_string_set("Enter the first set of strings, separated by commas: ")
    set2 = get_string_set("Enter the second set of strings, separated by commas: ")
    
    # Find the difference between the two sets (elements in set1 but not in set2)
    difference = set1 - set2
    
    # Print the elements that are in the first set but not in the second set
    print("Elements present in the first set but not in the second set:", difference)

# Run the program
main()
```

```python
def find_min_max(tpl):
    # Initialize min and max with the first element of the tuple
    min_val, max_val = tpl[0], tpl[0]
    
    # Iterate through the tuple to find the actual min and max
    for num in tpl:
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num
    
    # Return the minimum and maximum values as a tuple
    return min_val, max_val

# Example usage
numbers = (3, 1, 4, 1, 5, 9, 2, 6, 5)
min_value, max_value = find_min_max(numbers)
print(f"Minimum value: {min_value}, Maximum value: {max_value}")
```

```python
def word_frequencies(word_list):
    # Initialize an empty dictionary to store word frequencies
    frequency_dict = {}
    
    # Iterate over each word in the list
    for word in word_list:
        if word in frequency_dict:
            # Increment the count for the word if it's already in the dictionary
            frequency_dict[word] += 1
        else:
            # Add the word to the dictionary with a count of 1 if it's not already present
            frequency_dict[word] = 1
    
    return frequency_dict

# Example usage
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
result = word_frequencies(words)
print(result)
```

```python
def merge_dicts(dict1, dict2):
    # Create a new dictionary to store the merged results
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary

    # Iterate over each key-value pair in the second dictionary
    for key, value in dict2.items():
        if key in merged_dict:
            # If the key is already in the merged dictionary, add the values together
            merged_dict[key] += value
        else:
            # If the key is not in the merged dictionary, add the key-value pair
            merged_dict[key] = value

    return merged_dict

# Example usage
dict1 = {'apple': 3, 'banana': 2}
dict2 = {'banana': 1, 'orange': 4}
result = merge_dicts(dict1, dict2)
print(result)
```

```python
def get_values_from_dict(dictionary, keys):
    # Create a list to store the corresponding values
    result = []
    
    # Iterate through each key in the list of keys
    for key in keys:
        # Append the value if the key exists, otherwise append None
        result.append(dictionary.get(key, None))
    
    return result

# Example usage
my_dict = {
    'a': 1,
    'b': 2,
    'c': 3
}
keys = ['a', 'b', 'x', 'c', 'y']
```

```python
def sort_dict_by_value(dictionary, ascending=True):
    # Sort the dictionary by its values
    sorted_items = sorted(dictionary.items(), key=lambda item: item[1], reverse=not ascending)
    
    # Convert the sorted items back to a dictionary
    sorted_dict = dict(sorted_items)
    
    return sorted_dict

# Example usage
my_dict = {
    'a': 3,
    'b': 1,
    'c': 2
}

# Sort in ascending order
sorted_dict_asc = sort_dict_by_value(my_dict, ascending=True)
print("Sorted in ascending order:", sorted_dict_asc)  # Output: {'b': 1, 'c': 2, 'a': 3}

# Sort in descending order
sorted_dict_desc = sort_dict_by_value(my_dict, ascending=False)
print("Sorted in descending order:", sorted_dict_desc)  # Output: {'a': 3, 'c': 2, 'b': 1}
```

```python
def invert_dict(dictionary):
    inverted_dict = {}
    
    for key, value in dictionary.items():
        if value in inverted_dict:
            # Append the key to the list of keys for this value
            inverted_dict[value].append(key)
        else:
            # Create a new list with this key for the value
            inverted_dict[value] = [key]
    
    return inverted_dict

# Example usage
original_dict = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3
}

inverted_dict = invert_dict(original_dict)
print("Inverted dictionary:", inverted_dict)
# Output: Inverted dictionary: {1: ['a', 'c'], 2: ['b'], 3: ['d']}
```
